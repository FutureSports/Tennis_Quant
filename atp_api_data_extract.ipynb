```python
all_items = []
next_token = None
page = 1

while True:
    graphql_query = {
        "query": """
        query GetPlayers($limit: Int!, $nextToken: String) {
          players(limit: $limit, nextToken: $nextToken) {
            items {
              active
              age
              associationId
              backhandStyle
              birthCity
              birthDate
              careerPrizeMoney
              coach
              doublesCareerLosses
              doublesCareerTitles
              doublesCareerWins
              doublesHighestRank
              doublesHighestRankDate
              doublesLossesYearToDate
              doublesRank
              doublesTitlesYearToDate
              doublesWinsYearToDate
              firstName
              handedness
              heightCm
              heightFt
              id
              lastName
              nationality
              natlId
              playerHeadshotUrl
              residence
              singlesCareerLosses
              singlesCareerTitles
              singlesCareerWins
              singlesHighestRank
              singlesHighestRankDate
              singlesLossesYearToDate
              singlesRank
              singlesTitlesYearToDate
              singlesWinsYearToDate
              weightKg
              weightLb
              yearToDatePrizeMoney
              yearTurnedPro
            }
            nextToken
          }
        }
        """,
        "variables": {
            "limit": 300,
            "nextToken": next_token
        }
    }

    print(f"Fetching page {page} with nextToken: {next_token}")
    response = requests.post(url, headers=headers, json=graphql_query)
    data = response.json()

    if "errors" in data:
        print("GraphQL error:", data["errors"])
        break

    players_data = data.get("data", {}).get("players", {})
    items = players_data.get("items", [])

    print(f"Fetched {len(items)} players on page {page}")
    all_items.extend(items)

    next_token = players_data.get("nextToken")
    if not next_token:
        print("No more pages.")
        break

    page += 1
    time.sleep(0.1)  # Be respectful of the API rate limit
```
